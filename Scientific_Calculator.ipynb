```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Function to perform symbolic calculations using sympy
def symbolic_calculator():
    print("\nEnter an expression to solve or differentiate:")
    expr = input("Expression: ")

    try:
        # Replace implicit multiplication (e.g., 3x -> 3*x)
        expr = expr.replace('x', '*x')
        expr = expr.replace('=', '-(') + ')'  # Convert '3*x + 2 = 0' to '3*x + 2 - (0)'

        # Parse and evaluate the expression
        expression = sp.sympify(expr)

        print("\nChoose an option:")
        print("1. Simplify")
        print("2. Differentiate")
        print("3. Integrate")
        print("4. Solve Equation")

        option = input("Option (1/2/3/4): ")

        if option == '1':
            result = sp.simplify(expression)
        elif option == '2':
            var = input("Differentiate with respect to: ")
            result = sp.diff(expression, var)
        elif option == '3':
            var = input("Integrate with respect to: ")
            result = sp.integrate(expression, var)
        elif option == '4':
            var = sp.symbols('x')
            result = sp.solve(expression, var)
        else:
            print("Invalid option!")
            return

        print(f"\nResult: {result}")

    except Exception as e:
        print(f"Error: {e}")

# Function to plot a graph using matplotlib
def plot_function():
    expr = input("\nEnter a function of x to plot (e.g., sin(x), x**2 + 2): ")

    try:
        # Define the function to plot
        x = sp.symbols('x')
        func = sp.sympify(expr)

        # Convert sympy expression to a numerical function
        f_lambdified = sp.lambdify(x, func, 'numpy')

        # Generate x values and corresponding y values
        x_vals = np.linspace(-10, 10, 400)
        y_vals = f_lambdified(x_vals)

        # Plot the function
        plt.plot(x_vals, y_vals, label=f"f(x) = {expr}")
        plt.title(f"Graph of f(x) = {expr}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.grid(True)
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Main loop for the calculator
def calculator():
    while True:
        print("\n--- Scientific Graphical Calculator ---")
        print("1. Symbolic Calculator (Simplify, Differentiate, Integrate, Solve)")
        print("2. Plot Function")
        print("3. Exit")

        choice = input("Choose an option (1/2/3): ")

        if choice == '1':
            symbolic_calculator()
        elif choice == '2':
            plot_function()
        elif choice == '3':
            print("Exiting calculator...")
            break
        else:
            print("Invalid option! Please try again.")

# Run the calculator
calculator()
```

```python
import streamlit as st
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Function to perform symbolic calculations using sympy
def symbolic_calculator(expr, option, var=None):
    try:
        expr = expr.replace('x', '*x')
        if option == 'Solve Equation':
            expr = expr.replace('=', '-(') + ')'  # Convert '3*x + 2 = 0' to '3*x + 2 - (0)'

        # Parse the expression
        expression = sp.sympify(expr)

        if option == 'Simplify':
            result = sp.simplify(expression)
        elif option == 'Differentiate':
            result = sp.diff(expression, var)
        elif option == 'Integrate':
            result = sp.integrate(expression, var)
        elif option == 'Solve Equation':
            result = sp.solve(expression, var)
        else:
            result = "Invalid option"

        return result

    except Exception as e:
        return f"Error: {e}"

# Function to plot a graph using matplotlib
def plot_function(expr):
    try:
        # Define the function to plot
        x = sp.symbols('x')
        func = sp.sympify(expr)

        # Convert sympy expression to a numerical function
        f_lambdified = sp.lambdify(x, func, 'numpy')

        # Generate x values and corresponding y values
        x_vals = np.linspace(-10, 10, 400)
        y_vals = f_lambdified(x_vals)

        # Create the plot
        plt.plot(x_vals, y_vals, label=f"f(x) = {expr}")
        plt.title(f"Graph of f(x) = {expr}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.grid(True)
        plt.legend()

        # Show the plot using Streamlit
        st.pyplot(plt)

    except Exception as e:
        st.write(f"Error: {e}")

# Streamlit App UI
def main():
    st.title("Scientific Graphical Calculator")

    # Sidebar for calculator type
    st.sidebar.title("Options")
    calc_type = st.sidebar.radio("Choose Calculator Type", ("Symbolic Calculator", "Plot Function"))

    if calc_type == "Symbolic Calculator":
        st.subheader("Symbolic Calculator")

        expr = st.text_input("Enter an expression:")
        option = st.selectbox("Choose an operation", ["Simplify", "Differentiate", "Integrate", "Solve Equation"])

        if option in ["Differentiate", "Integrate", "Solve Equation"]:
            var = st.text_input("Enter the variable:", "x")
        else:
            var = None

        if st.button("Calculate"):
            result = symbolic_calculator(expr, option, var)
            st.write(f"Result: {result}")

    elif calc_type == "Plot Function":
        st.subheader("Plot a Mathematical Function")

        expr = st.text_input("Enter a function of x (e.g., sin(x), x**2 + 2):")

        if st.button("Plot"):
            plot_function(expr)

# Run the Streamlit app
if __name__ == "__main__":
    main()
```
